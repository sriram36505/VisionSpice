```python
import torch
from src.dataset_loader import load_dataset
from src.cnn_model import CNNModel
from src.train_utils import train_epoch, eval_epoch

device = "cuda" if torch.cuda.is_available() else "cpu"
train_loader, val_loader, test_loader, classes = load_dataset("/path/to/indian_spices")

model = CNNModel(len(classes)).to(device)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(10):
    train_loss, train_acc = train_epoch(model, train_loader, optimizer, criterion, device)
    val_loss, val_acc = eval_epoch(model, val_loader, criterion, device)
    print(f"Epoch {epoch+1}: val acc {val_acc:.3f}")
```
